```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# r = np.load('models/test/rewards_train.npy')
r = np.load('models/res_newh32/rewards_train.npy')
plt.plot(r)
plt.show()
```

```python
def moving_average(data_set, periods=3):
    weights = np.ones(periods) / periods
    return np.convolve(data_set, weights, mode='valid')
```

```python
plt.figure(figsize=(15,10))
plt.plot(r[:300], color='peachpuff')
plt.plot(moving_average(r, 20)[:300])
plt.title('Running training reward - Oval track', fontsize=15)
plt.xlabel('Episodes', fontsize=15)
plt.ylabel('Reward value', fontsize=15)
```
