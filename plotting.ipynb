```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# r = np.load('models/test/rewards_train.npy')
r = np.load('models/res_newh32/rewards_train.npy')
plt.plot(r)
plt.show()
```

```python
def moving_average(data_set, periods=3):
    weights = np.ones(periods) / periods
    return np.convolve(data_set, weights, mode='valid')
```

```python
plt.figure(figsize=(15,10))
plt.plot(r[:300], color='peachpuff')
plt.plot(moving_average(r, 20)[:300])
plt.title('Training with features extracted from visual input', fontsize=20)
plt.xlabel('Episodes', fontsize=20)
plt.ylabel('Reward value', fontsize=20)
```

```python
r_b = np.load('models/base_rewards_train.npy')
plt.plot(r_b)
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.plot(r_b[:475], color='peachpuff')
plt.plot(moving_average(r_b, 20)[:475])
plt.title('Training with only low level features', fontsize=20)
plt.xlabel('Episodes', fontsize=20)
plt.ylabel('Reward value', fontsize=20)
```
